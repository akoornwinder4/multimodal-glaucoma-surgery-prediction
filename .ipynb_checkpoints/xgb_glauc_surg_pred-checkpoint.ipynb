```python
# F1 tuning function
def f1_tuning(test_pred,y_val):
    f1scores={}
    for i in np.arange(0.05,1.0,0.05):
        new_pred = np.where(test_pred<i, 0 , 1)
        f1scores[i]=f1_score(y_val,new_pred)
        
    return f1scores,max(f1scores, key=f1scores.get)
```

## Import Data

```python
# RNFL data
rnfl_train = pd.read_csv('data_updated/rnfl_train_arvo.csv')
rnfl_val = pd.read_csv('data_updated/rnfl_val_arvo.csv')
rnfl_test = pd.read_csv('data_updated/rnfl_test_arvo.csv')

# RNFL w/ EHR 
rnflehr_train = pd.read_csv('data_updated/rnflehr_train_arvo.csv')
rnflehr_val = pd.read_csv('data_updated/rnflehr_val_arvo.csv')
rnflehr_test = pd.read_csv('data_updated/rnflehr_test_arvo.csv')

# Only EHR 
ehr_train = pd.read_csv('data_updated/ehr_train_arvo.csv')
ehr_val = pd.read_csv('data_updated/ehr_val_arvo.csv')
ehr_test = pd.read_csv('data_updated/ehr_test_arvo.csv')
```

```python
X_train = rnfl_train.drop(columns=['target', 'pat_mrn'])
X_val = rnfl_val.drop(columns=['target', 'pat_mrn'])
X_test = rnfl_test.drop(columns=['target', 'pat_mrn'])

y_train = rnfl_train['target']
y_val = rnfl_val['target']
y_test = rnfl_test['target']
```

## SMOTE

```python
# Apply SMOTE to the training set to balance class distribution
smote = SMOTE(random_state=42)  
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

## XGB

```python
xgb_clf = xgb.XGBClassifier()
xgb_clf.fit(X_train_resampled, y_train_resampled)

print(xgb_clf)

expected_y  = y_val
predicted_y = xgb_clf.predict(X_val)

print(metrics.classification_report(expected_y, predicted_y))
print(metrics.confusion_matrix(expected_y, predicted_y))

print('Valid',roc_auc_score(expected_y,predicted_y))
```

```python
# Hyperparameter tuning
param_grid_xgb = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}

# Perform hyperparameter tuning for the model using the validation set
grid_search_xgb = GridSearchCV(xgb_clf, param_grid_xgb, cv=5, scoring='f1', n_jobs=-1)
grid_search_xgb.fit(X_val, y_val)

best_xgb_model = grid_search_xgb.best_estimator_

y_pred_xgb = best_xgb_model.predict(X_val)

yval_probs_xgb = best_xgb_model.predict_proba(X_val)
print('XGBoost validation roc-auc: ', roc_auc_score(y_val, yval_probs_xgb[:,1]))
print()

report_xgb = classification_report(y_val, y_pred_xgb)
print("Classification Report for XGBoost on Validation Set:")
print(report_xgb)

best_params_xgb = grid_search_xgb.best_params_
print("Best Hyperparameters for XGBoost:", best_params_xgb)
```

## F1 Thresholding

```python
best_threshold = 0.5  # Starting threshold

best_f1_score = 0
optimal_threshold = best_threshold

for threshold in [0.3, 0.4, 0.5, 0.6, 0.7]:
    y_pred_thresholded = (yval_probs_xgb[:, 1] > threshold).astype(int)
    f1 = f1_score(y_val, y_pred_thresholded)
    
    if f1 > best_f1_score:
        best_f1_score = f1
        optimal_threshold = threshold

print("Optimal Threshold:", optimal_threshold)
print("Best F1 Score:", best_f1_score)
```

## Eval on test set

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, roc_auc_score, roc_curve

# XGBoost model
model = best_xgb_model  # Replace with your XGBoost model
model_name = "XGBoost"
dataset_name = "RNFL only"
optimal_threshold = 0.5  # Replace with your optimal threshold - DONT FORGET THIS

# Predict probabilities for the positive class on the test set
y_pred_probs_test = model.predict_proba(X_test)[:, 1]

# Calculate ROC AUC score
roc_auc = roc_auc_score(y_test, y_pred_probs_test)

# Generate classification report
y_pred_thresholded = (y_pred_probs_test > optimal_threshold).astype(int)
report = classification_report(y_test, y_pred_thresholded, digits=3)
print(f"Classification Report for {model_name} on {dataset_name} Test Set:")
print(report)

# Calculate ROC curve
fpr, tpr, _ = roc_curve(y_test, y_pred_probs_test)

# Plot ROC curve for the XGBoost model
plt.figure()
plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(f'Receiver Operating Characteristic - {model_name}')
plt.legend(loc='lower right')
plt.show()
```

```python
# Save model

import joblib

# Assuming you have your XGBoost model stored in the 'best_xgb_model' variable
model = best_xgb_model  # Replace with your XGBoost model

# Define the file path where you want to save the model
model_filename = 'models_updated/xgboost_rnfl_glauc_model.pkl'

# Save the model to the specified file using joblib
joblib.dump(model, model_filename)

print(f"Model saved to {model_filename}")
```

## Plot AUROC Curves To Compare

### Load Models

```python
rnfl_clf = joblib.load("models_updated/xgboost/xgboost_rnfl_glauc_model.pkl")
ehr_clf = joblib.load("models_updated/xgboost/xgboost_ehr_glauc_model.pkl")
fusion_clf = joblib.load("models_updated/xgboost/xgboost_fusion_glauc_model.pkl")
```

### Load Test Set Data

```python
rnfl_test = pd.read_csv('data_updated/rnfl_test_arvo.csv')
rnflehr_test = pd.read_csv('data_updated/rnflehr_test_arvo.csv')
ehr_test = pd.read_csv('data_updated/ehr_test_arvo.csv')

X_rnfl_test = rnfl_test.drop(columns=['pat_mrn', 'target'])
y_rnfl_test = rnfl_test.target

X_fusion_test = rnflehr_test.drop(columns=['pat_mrn', 'target'])
y_fusion_test = rnflehr_test.target

X_ehr_test = ehr_test.drop(columns=['pat_mrn', 'target'])
y_ehr_test = ehr_test.target
```

## Make Predictions and Plot AUROC Curves

```python
# Make predictions using the models
preds_rnfl = rnfl_clf.predict_proba(X_rnfl_test)[:, 1]
preds_fusion = fusion_clf.predict_proba(X_fusion_test)[:, 1]
preds_ehr = ehr_clf.predict_proba(X_ehr_test)[:, 1]

# Compute ROC curve and AUC for each model
fpr_rnfl, tpr_rnfl, _ = roc_curve(y_rnfl_test, preds_rnfl)
roc_auc_rnfl = auc(fpr_rnfl, tpr_rnfl)

fpr_fusion, tpr_fusion, _ = roc_curve(y_fusion_test, preds_fusion)
roc_auc_fusion = auc(fpr_fusion, tpr_fusion)

fpr_ehr, tpr_ehr, _ = roc_curve(y_ehr_test, preds_ehr)
roc_auc_ehr = auc(fpr_ehr, tpr_ehr)

# Plot the ROC curves
plt.figure(figsize=(5, 5))
plt.plot(fpr_fusion, tpr_fusion, color='green', lw=2, label=f'Fusion Model (AUC = {roc_auc_fusion:.3f})')
plt.plot(fpr_ehr, tpr_ehr, color='red', lw=2, label=f'EHR Model (AUC = {roc_auc_ehr:.3f})')
plt.plot(fpr_rnfl, tpr_rnfl, color='blue', lw=2, label=f'RNFL Model (AUC = {roc_auc_rnfl:.3f})')
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('XGBoost Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.savefig('plots_updated/xgb/xgb_roc_curves.png', bbox_inches='tight')  # Save the RNFL plot
plt.show()
```

## AUCPRC Curves

```python
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt

# Compute Precision-Recall curve and AUC for each model
precision_rnfl, recall_rnfl, _ = precision_recall_curve(y_rnfl_test, preds_rnfl)
prc_auc_rnfl = auc(recall_rnfl, precision_rnfl)

precision_fusion, recall_fusion, _ = precision_recall_curve(y_fusion_test, preds_fusion)
prc_auc_fusion = auc(recall_fusion, precision_fusion)

precision_ehr, recall_ehr, _ = precision_recall_curve(y_ehr_test, preds_ehr)
prc_auc_ehr = auc(recall_ehr, precision_ehr)

# Plot the Precision-Recall curves
plt.figure(figsize=(5, 5))
plt.plot(recall_fusion, precision_fusion, color='green', lw=2, label=f'Fusion Model (AUC-PRC = {prc_auc_fusion:.3f})')
plt.plot(recall_ehr, precision_ehr, color='red', lw=2, label=f'EHR Model (AUC-PRC = {prc_auc_ehr:.3f})')
plt.plot(recall_rnfl, precision_rnfl, color='blue', lw=2, label=f'RNFL Model (AUC-PRC = {prc_auc_rnfl:.3f})')
plt.xlabel('Recall (Sensitivity)')
plt.ylabel('Precision (Positive Predictive Value)')
plt.title('XGBoost Precision-Recall Curves')
plt.legend(loc='lower left')
plt.savefig('plots_updated/xgb/xgb_prc_curves.png', bbox_inches='tight')  # Save the PRC plot
plt.show()
```

```python
import shap
import xgboost as xgb
import matplotlib.pyplot as plt

# Load your XGBoost models (rnfl_clf, fusion_clf, ehr_clf) here
rnfl_model = joblib.load("models_updated/xgboost/xgboost_rnfl_glauc_model.pkl")
ehr_model = joblib.load("models_updated/xgboost/xgboost_ehr_glauc_model.pkl")
fusion_model = joblib.load("models_updated/xgboost/xgboost_fusion_glauc_model.pkl")


# Define your test data (X_rnfl_test, X_fusion_test, X_ehr_test) here
rnfl_test = pd.read_csv('data_updated/rnfl_test_arvo.csv')
rnflehr_test = pd.read_csv('data_updated/rnflehr_test_arvo.csv')
ehr_test = pd.read_csv('data_updated/ehr_test_arvo.csv')

X_rnfl_test = rnfl_test.drop(columns=['pat_mrn', 'target'])
y_rnfl_test = rnfl_test.target

X_fusion_test = rnflehr_test.drop(columns=['pat_mrn', 'target'])
y_fusion_test = rnflehr_test.target

X_ehr_test = ehr_test.drop(columns=['pat_mrn', 'target'])
y_ehr_test = ehr_test.target



# Initialize the explainer for each model
rnfl_explainer = shap.Explainer(rnfl_model)
fusion_explainer = shap.Explainer(fusion_model)
ehr_explainer = shap.Explainer(ehr_model)



# Calculate Shapley values for the test data
rnfl_shap_values = rnfl_explainer.shap_values(X_rnfl_test)
fusion_shap_values = fusion_explainer.shap_values(X_fusion_test)
ehr_shap_values = ehr_explainer.shap_values(X_ehr_test)



# Summarize Shapley values for each model
shap.summary_plot(rnfl_shap_values, X_rnfl_test, feature_names=X_rnfl_test.columns,
                  show=False, plot_type='bar', max_display=20)
plt.title('Shapley Summary Plot - XGB RNFL Model')
plt.savefig('plots_updated/xgb/xgb_rnfl_shap_bar_plot.png', bbox_inches='tight')  # Save the RNFL plot
plt.show()

shap.summary_plot(fusion_shap_values, X_fusion_test, feature_names=X_fusion_test.columns,
                  show=False, plot_type='bar', max_display=20)
plt.title('Shapley Summary Plot - XGB Fusion Model')
plt.savefig('plots_updated/xgb/xgb_fusion_shap_bar_plot.png', bbox_inches='tight')  # Save the RNFL plot
plt.show()

shap.summary_plot(ehr_shap_values, X_ehr_test, feature_names=X_ehr_test.columns,
                  show=False, plot_type='bar', max_display=20)
plt.title('Shapley Summary Plot - XGB EHR Model')
plt.savefig('plots_updated/xgb/xgb_ehr_shap_bar_plot.png', bbox_inches='tight')  # Save the RNFL plot
plt.show()
```

```python
import shap
import pandas as pd
import joblib
import xgboost as xgb
import matplotlib.pyplot as plt

# Load your XGBoost models (rnfl_clf, fusion_clf, ehr_clf) here
rnfl_model = joblib.load("models_updated/xgboost/xgboost_rnfl_glauc_model.pkl")
ehr_model = joblib.load("models_updated/xgboost/xgboost_ehr_glauc_model.pkl")
fusion_model = joblib.load("models_updated/xgboost/xgboost_fusion_glauc_model.pkl")


# Define your test data (X_rnfl_test, X_fusion_test, X_ehr_test) here
rnfl_test = pd.read_csv('data_updated/rnfl_test_arvo.csv')
rnflehr_test = pd.read_csv('data_updated/rnflehr_test_arvo.csv')
ehr_test = pd.read_csv('data_updated/ehr_test_arvo.csv')

X_rnfl_test = rnfl_test.drop(columns=['pat_mrn', 'target'])
y_rnfl_test = rnfl_test.target

X_fusion_test = rnflehr_test.drop(columns=['pat_mrn', 'target'])
y_fusion_test = rnflehr_test.target

X_ehr_test = ehr_test.drop(columns=['pat_mrn', 'target'])
y_ehr_test = ehr_test.target



# Initialize the explainer for each model
rnfl_explainer = shap.Explainer(rnfl_model)
fusion_explainer = shap.Explainer(fusion_model)
ehr_explainer = shap.Explainer(ehr_model)



# Calculate Shapley values for the test data
rnfl_shap_values = rnfl_explainer.shap_values(X_rnfl_test)
fusion_shap_values = fusion_explainer.shap_values(X_fusion_test)
ehr_shap_values = ehr_explainer.shap_values(X_ehr_test)



# Summarize Shapley values for each model
shap.summary_plot(rnfl_shap_values, X_rnfl_test, feature_names=X_rnfl_test.columns,
                  show=False, plot_type='dot', max_display=20)
plt.title('Shapley Summary Plot - XGB RNFL Model')
# plt.savefig('plots_updated/xgb/xgb_rnfl_shap_dot_plot.png', bbox_inches='tight')  # Save the RNFL plot
plt.show()

shap.summary_plot(fusion_shap_values, X_fusion_test, feature_names=X_fusion_test.columns,
                  show=False, plot_type='dot', max_display=20)
plt.title('Shapley Summary Plot - XGB Fusion Model')
# plt.savefig('plots_updated/xgb/xgb_fusion_shap_dot_plot.png', bbox_inches='tight')  # Save the RNFL plot
plt.show()

shap.summary_plot(ehr_shap_values, X_ehr_test, feature_names=X_ehr_test.columns,
                  show=False, plot_type='dot', max_display=20)
plt.title('Shapley Summary Plot - XGB EHR Model')
# plt.savefig('plots_updated/xgb/xgb_ehr_shap_dot_plot.png', bbox_inches='tight')  # Save the RNFL plot
plt.show()
```
